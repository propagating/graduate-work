```python
cmu = nltk.corpus.cmudict.entries()

orange_syllables =[(word, sound) for word, sound in cmu if word == 'orange']
orange_rhymes = []
for (word, syllable) in orange_syllables:
         orange_rhymes += [word for word, pron in cmu if pron[-1:] == syllable[-1:]]

print(set(orange_rhymes))

# The output will change depending on the 'level' of rhyming that you want, the smaller (because we are looking at negative numbers) the value you use, the smaller the list will be, but they will rhyme more closely.
```

```python
seven_syllables =[(word, sound) for word, sound in cmu if word == 'seven']
seven_rhymes = []
for (word, syllable) in seven_syllables:
         seven_rhymes += [word for word, pron in cmu if pron[-1:] == syllable[-1:]]

print(set(seven_rhymes))

# We see many more words that rhyme with seven at the same 'rhyme level' as orange
```

# Part and Substance Holonyms

According to word net, a Holonym is a relationship between 2 concepts, where concept A makes up a part of Concept B. The example give is that 'eye' is a part_holonym of face. A substance holonym is somewhat similar, in that hydrogen would be a substance holonym of water, as hydrogen is part of the substance that makes up water.

When we look at the part holonyms for mint.n.04, we see that mint.n.02 is the only part_holonym, meaning that mint.n.04 is a part of mint.n.02. Simialrly, for the substance holonyms we get mint.n.05, meaning that mint.n.04 is part of the substance that makes up mint.n.05.

```python
def convert_to_pig_latin(word):
    vowels = ['a', 'e', 'i', 'o', 'u', 'y']
    graphemes = ['b', 'bb', 'd', 'dd', 'ed', 'f', 'ff', 'ph', 'gh', 'lf', 'ft', 'g', 'gg', 'gh', 'gu', 'gue', 'h', 'wh',
                 'j', 'ge', 'g', 'dge', 'di', 'gg', 'k', 'c', 'ch', 'cc', 'lk', 'qu', 'ck', 'x', 'l', 'll', 'm', 'mm',
                 'mb', 'mn', 'lm', 'n', 'nn', 'kn', 'gn', 'pn', 'mn', 'p', 'pp', 'r', 'rr', 'wr', 'rh', 's', 'ss', 'c',
                 'sc', 'ps', 'st', 'ce', 'se', 't', 'tt', 'th', 'ed', 'v', 'f', 'ph', 've', 'w', 'wh', 'u', 'o', 'z',
                 'zz', 's', 'ss', 'x', 'ze', 'se', 's', 'si', 'z', 'ch', 'tch', 'tu', 'te', 'sh', 'ce', 's', 'ci', 'si',
                 'ch', 'sci', 'ti', 'th', 'th', 'ng', 'n', 'ngue', 'y', 'i', 'j', ]

    pig_graphs = set([g for g in graphemes if len(g) > 1])
    pig_word = ''

    # if first letter is a vowel

    if word[0] in vowels:
        pig_word += word
        pig_word += 'ay'
        return pig_word

    for g in graphemes:
        if word.startswith(g):
            g_len = len(g)
            pig_word += word[g_len:]
            pig_word += word[:g_len]
            pig_word += 'ay'
            return pig_word

    # we want to look at the first phoneme of the word for cases like 'qu', 'th'
    for i in range(len(word)):
        if word[i] in vowels:
            pig_word += word[i:]
            pig_word += word[:i]
            pig_word += 'ay'
            return pig_word

        # if we get here the word has no vowels
    pig_word += word
    pig_word += 'ay'
    return pig_word
```

```python
convert_to_pig_latin('eat')

convert_to_pig_latin('hello')

convert_to_pig_latin('quick')
```

# Brown Hobbies Vs Shakespeare

In the Brown Corpus, with is generally used to indicate some form of ownership over an object, idea, or in the description of an object or individual. It is rarely, if ever, used to accompany an action that is taking place, which is how Shakespeare tends to use the word. It's the difference between describing a person : 'What are you going to do with your Muscles?', and describing an action : 'what's the news with you?', 'almost to doomsday with eclipse'.

I also find that the hobbies genre tends to use the possessive form of 'with' to describe physical objects or attributes of a person, where Shakespeare tends to use the possessive form of the word more towards the intangible things such as emotions and time. Maybe it was an odd choice to compare these two corpora, as it's a bit difficult to fully express their differences, being that Shakespeare is writing a play for his audience, his wording and dialogue is used to try and convey an atmosphere over an audience, and when used as part of a Character's speech then he uses it to try and display emotions through in that speech such that the audience can feel that emotion come through.

Compared to the hobbies text, it's much more grounded in reality and instead of trying to show us an emotion or idea of a place or thing, it's trying to concretely tell us what something or someone looks like, and how that person or object inhabits their current space.
